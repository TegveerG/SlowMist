# Web Scraper for https://hacked.slowmist.io/?c=&page=1

```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
# scrape the data from the website 
# url - https://hacked.slowmist.io/?c=&page=1
# total pages = 56

# create a list of urls to scrape
urls = []
for i in range(1,57):
    urls.append('https://hacked.slowmist.io/?c=&page='+str(i))

# each record is stored in website as <li> tag
```

```python
# URLs
base_url = 'https://hacked.slowmist.io/?c=&page='
num_pages = 56

# Initialize empty lists for storing data
time_list = []
hacked_target_list = []
description_list = []
funds_lost_list = []
attack_method_list = []
source_list = []

# Scrape data from each page
for page in range(1, num_pages + 1):
    url = base_url + str(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all <li> elements
    records = soup.find('div', class_='case-content')
    #print(records)
    # Check if records exist
    if records:
        # Find all <li> elements
        attacks = records.find_all('li')
        # Extract data from each record
        for attack in attacks:
            time = attack.find('span', class_='time').text.strip()
            print(time)
            hacked_target = attack.find('h3').text.strip()
            print(hacked_target)
            description = attack.find('p').text.strip()
            print(description)
            #funds_lost = attack.find('span').text.strip()
            funds_lost = attack.find('em').find_next('span').text.strip()
            print(funds_lost)
            attack_method = attack.find('em').find_next('span').text.strip()
            print(attack_method)
            source = attack.find('a', text='View Reference Sources')['href']
            print(source)

            # Append data to respective lists
            time_list.append(time)
            hacked_target_list.append(hacked_target)
            description_list.append(description)
            funds_lost_list.append(funds_lost)
            attack_method_list.append(attack_method)
            source_list.append(source)

# Create a pandas DataFrame
data = {
    'Date': time_list,
    'Hacked_Target': hacked_target_list,
    'Description': description_list,
    'Funds_Lost': funds_lost_list,
    'Attack_Method': attack_method_list,
    'Source': source_list
}

df = pd.DataFrame(data)
```

```python
# URLs
base_url = 'https://hacked.slowmist.io/?c=&page='
num_pages = 56

# Initialize empty lists for storing data
time_list = []
hacked_target_list = []
description_list = []
funds_lost_list = []
attack_method_list = []
source_list = []

# Scrape data from each page
for page in range(1, num_pages + 1):
    url = base_url + str(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all <li> elements
    #records = soup.find_all('li')
    
    # Extract data from each record
    for record in soup.find_all('li'):
        time = record.find('span', class_='time').replace(' ', '')
        print(time)
        hacked_target = record.find('em', string='Hacked target:').find_next('h3')
        description = record.find('em', string='Description of the event:').find_next('p')
        funds_lost = record.find('em', string='Amount of loss:').find_next('span')
        attack_method = record.find('em', string='Attack method:').find_next('span')
        source = record.find('a', text='View Reference Sources')

        # Check if elements exist before accessing the text attribute
        if time:
            time = time.text.strip()
        else:
            time = ''
        print(time)
        if hacked_target:
            hacked_target = hacked_target.text.strip()
        else:
            hacked_target = ''
        print(hacked_target)
        if description:
            description = description.text.strip()
        else:
            description = ''
        print(description)
        if funds_lost:
            funds_lost = funds_lost.text.strip()
        else:
            funds_lost = ''
        print(funds_lost)
        if attack_method:
            attack_method = attack_method.text.strip()
        else:
            attack_method = ''
        print(attack_method)
        if source:
            source = source['href']
        else:
            source = ''
        print(source)
        # Append data to respective lists
        time_list.append(time)
        hacked_target_list.append(hacked_target)
        description_list.append(description)
        funds_lost_list.append(funds_lost)
        attack_method_list.append(attack_method)
        source_list.append(source)

# Create a pandas DataFrame
data = {
    'Time': time_list,
    'Hacked_Target': hacked_target_list,
    'Description': description_list,
    'Funds_Lost': funds_lost_list,
    'Attack_Method': attack_method_list,
    'Source': source_list
}

df = pd.DataFrame(data)
```
